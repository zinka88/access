```python
# This query represents dataset "breast_cancer" for domain "condition" and was generated for All of Us Controlled Tier Dataset v7
dataset_14558003_condition_sql = """
    SELECT
        c_occurrence.person_id,
        c_occurrence.condition_concept_id,
        c_occurrence.condition_start_date
    FROM
        ( SELECT
            * 
        FROM
            `""" + os.environ["WORKSPACE_CDR"] + """.condition_occurrence` c_occurrence 
        WHERE
            (
                condition_concept_id IN (SELECT
                    DISTINCT c.concept_id 
                FROM
                    `""" + os.environ["WORKSPACE_CDR"] + """.cb_criteria` c 
                JOIN
                    (SELECT
                        CAST(cr.id as string) AS id       
                    FROM
                        `""" + os.environ["WORKSPACE_CDR"] + """.cb_criteria` cr       
                    WHERE
                        concept_id IN (4112853)       
                        AND full_text LIKE '%_rank1]%'      ) a 
                        ON (c.path LIKE CONCAT('%.', a.id, '.%') 
                        OR c.path LIKE CONCAT('%.', a.id) 
                        OR c.path LIKE CONCAT(a.id, '.%') 
                        OR c.path = a.id) 
                WHERE
                    is_standard = 1 
                    AND is_selectable = 1)
            )  
            AND (
                c_occurrence.PERSON_ID IN (SELECT
                    distinct person_id  
                FROM
                    `""" + os.environ["WORKSPACE_CDR"] + """.cb_search_person` cb_search_person  
                WHERE
                    cb_search_person.person_id IN (SELECT
                        criteria.person_id 
                    FROM
                        (SELECT
                            DISTINCT person_id, entry_date, concept_id 
                        FROM
                            `""" + os.environ["WORKSPACE_CDR"] + """.cb_search_all_events` 
                        WHERE
                            (concept_id IN(SELECT
                                DISTINCT c.concept_id 
                            FROM
                                `""" + os.environ["WORKSPACE_CDR"] + """.cb_criteria` c 
                            JOIN
                                (SELECT
                                    CAST(cr.id as string) AS id       
                                FROM
                                    `""" + os.environ["WORKSPACE_CDR"] + """.cb_criteria` cr       
                                WHERE
                                    concept_id IN (43528895)       
                                    AND full_text LIKE '%_rank1]%'      ) a 
                                    ON (c.path LIKE CONCAT('%.', a.id, '.%') 
                                    OR c.path LIKE CONCAT('%.', a.id) 
                                    OR c.path LIKE CONCAT(a.id, '.%') 
                                    OR c.path = a.id) 
                            WHERE
                                is_standard = 0 
                                AND is_selectable = 1) 
                            AND is_standard = 0 )) criteria ) )
            )) c_occurrence"""

breast_cancer = pd.read_gbq(
    dataset_14558003_condition_sql,
    dialect="standard",
    use_bqstorage_api=("BIGQUERY_STORAGE_API_ENABLED" in os.environ),
    progress_bar_type="tqdm_notebook")

breast_cancer['condition'] = 'breast cancer'
breast_cancer.head(5)
```

```python
# subset to rows without missing self survey date
surveydates = access_df.dropna(subset=['self_survey_date'])
```

```python
ehr_df = pd.DataFrame()

for condition in tqdm(condition_list, desc="Processing Conditions", leave=True):

    condition_df = globals()[condition]
    
    condition_df['condition'] = condition_df['condition'].str.replace(' ', '_', regex=True)
    df_merged = pd.merge(condition_df, surveydates)
    df_merged['self_survey_date'] = pd.to_datetime(df_merged['self_survey_date'])
    df_merged['condition_start_date'] = pd.to_datetime(df_merged['condition_start_date'])
    df_merged['diff_days'] = (df_merged['self_survey_date'] - df_merged['condition_start_date']).dt.days
    df_merged['ehr_ever'] = 1 
    df_merged['ehr_lookback'] = (df_merged['diff_days'] >= 0).astype(int)
    df_merged['ehr_1_yr_lookback'] = ((df_merged['diff_days'] >=0) & (df_merged['diff_days'] <=365)).astype(int)
    df_merged['ehr_2_yr_lookback'] = ((df_merged['diff_days'] >=0) & (df_merged['diff_days'] <=730)).astype(int)
    df_grouped = df_merged.groupby(['person_id','condition'], as_index=False)[['ehr_ever','ehr_lookback','ehr_1_yr_lookback','ehr_2_yr_lookback']].max()
    
    ehr_df = pd.concat([ehr_df, df_grouped], ignore_index=True)
    
```
